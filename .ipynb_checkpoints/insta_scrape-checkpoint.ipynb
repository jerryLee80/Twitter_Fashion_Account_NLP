```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.instagram.com/gucci/')
```

```python
sleep(2)
html = driver.page_source

soup = BeautifulSoup(html, 'lxml')
```

```python
caption = soup.find('li', {'class':'_ezgzd'}).text
caption
```

```python
button = driver.find_element_by_link_text('Next')
button.click()
```

```python
df = pd.DataFrame(columns=['Label', 'description'])

def scrapper(soup, df):
    for page in soup:
        
        #define variables
        label = soup.find('a',{'class':'_2g7d5 notranslate _iadoq'}).text
        caption = soup.find('li', {'class':'_ezgzd'}).text
        button = driver.find_element_by_link_text('Next')
        
        #save to dataframe 
        df.loc[len(df)] = [label, caption]
        
        #go to next picture
        button.click()
        
    return df
scrapper(soup, df)
```
